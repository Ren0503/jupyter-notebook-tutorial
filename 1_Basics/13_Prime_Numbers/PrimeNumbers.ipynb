# Prime Numbers 
### Số nguyên tố sử dụng Python
Số nguyên tố là gì
- Là một số tự nhiên (số nguyên dương).
- Chỉ chia hết cho hai số là 1 và chính nó.
Nhiệm vụ: Viết chương trình tạo danh sách số nguyên tố nhỏ hơn 500.
## Cách 1: Dùng For

```python
def approach1(givenNumber):
    
    # Tạo list
    primes = []
    
    for possiblePrime in range(2, givenNumber + 1):
        
        # Giả sử là số nguyên tố cho đến khi biết nó không phải.
        isPrime = True
        for num in range(2, possiblePrime):
            if possiblePrime % num == 0:
                isPrime = False
        
        if isPrime:
            primes.append(possiblePrime)
    return(primes)
```

```python
approach1(500)
```

## Cách 2: Dùng for với break
Chỉ có sự khác biệt là lệnh break. Về cơ bản khi bạn biết rằng một số không phải là số nguyên tố, tại sao phải tiếp tục kiểm tra nó trong vòng lặp for.

```python
def approach2(givenNumber):

    # Tạo list
    primes = []
    for possiblePrime in range(2, givenNumber):

        # Giả sử là số nguyên tố cho đến khi biết nó không phải.
        isPrime = True
        for num in range(2, possiblePrime):
            if possiblePrime % num == 0:
                isPrime = False
                break

        if isPrime:
            primes.append(possiblePrime)
    
    return(primes)
```

```python
approach2(500)
```

## Cách 3: Dùng For, Break và Căn bậc 2

```python
def approach3(givenNumber):  
    
    # Tạo list
    primes = []
    for possiblePrime in range(2, givenNumber + 1):

        # Giả sử là số nguyên tố cho đến khi biết nó không phải.
        isPrime = True
        for num in range(2, int(possiblePrime ** 0.5) + 1):
            if possiblePrime % num == 0:
                isPrime = False
                break

        if isPrime:
            primes.append(possiblePrime)
    
    return(primes)
```

```python
approach3(500)
```
